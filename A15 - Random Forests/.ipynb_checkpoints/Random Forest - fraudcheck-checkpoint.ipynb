# 1. Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

# 2. Importing the data

```python
fraudcheck = pd.read_csv('Fraud_check.csv')
fraudcheck
```

# 3. Data Understanding

```python
fraudcheck.shape
```

```python
fraudcheck.dtypes
```

```python
fraudcheck.describe()
```

```python
fraudcheck.isnull().sum()
```

# 4. Data Preprocessing

```python
fraudcheck.rename(columns={'Marital.Status':'Marital_Status','Taxable.Income':'Taxable_Income','City.Population':'City_Population','Work.Experience':'Work_Experience'},inplace= True)
fraudcheck
```

```python
fraudcheck['Taxable_Income_Status'] = fraudcheck.Taxable_Income.map(lambda x: 'Risky' if x<=30000 else 'Good')
```

```python
fraudcheck
```

```python
fraudcheck_label = LabelEncoder()
fraudcheck['Marital_Status'] = fraudcheck_label.fit_transform(fraudcheck['Marital_Status'])
fraudcheck['Undergrad'] = fraudcheck_label.fit_transform(fraudcheck['Undergrad'])
fraudcheck['Urban'] = fraudcheck_label.fit_transform(fraudcheck['Urban'])
fraudcheck['Taxable_Income_Status'] = fraudcheck_label.fit_transform(fraudcheck['Taxable_Income_Status'])
```

```python
fraudcheck
```

# 5. Model Building

```python
X = fraudcheck.drop(['Taxable_Income','Taxable_Income_Status'],axis=1)
y = fraudcheck[['Taxable_Income_Status']]
```

```python
X
```

```python
y
```

```python
X.shape,y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=50)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# 6. Model Training

```python
fraudcheck_RF = RandomForestClassifier(n_estimators=100,criterion='gini',max_depth=None)
fraudcheck_RF.fit(X_train,y_train)
```

# 7. Model Testing

```python
y_pred_train = fraudcheck_RF.predict(X_train)
y_pred_test = fraudcheck_RF.predict(X_test)
```

```python
y_pred_train
```

```python
y_pred_test
```

# 8. Model Evaluation

```python
print("Train Accuracy Score :",(accuracy_score(y_train,y_pred_train)))
print("Confusion Matric :\n",(confusion_matrix(y_train,y_pred_train)))
```

```python
print("Test Accuracy Score :",(accuracy_score(y_test,y_pred_test)))
print("Confusion Matric :\n",(confusion_matrix(y_test,y_pred_test)))
```
