# 1. Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

# 2. Importing the data

```python
company_data = pd.read_csv('Company_Data.csv')
company_data
```

# 3. Data Understanding

```python
company_data.shape
```

```python
company_data.dtypes
```

```python
company_data.describe()
```

```python
company_data.isnull().sum()
```

# 4. Data Preprocessing

```python
company_label = LabelEncoder()
company_data['ShelveLoc'] = company_label.fit_transform(company_data['ShelveLoc'])
company_data['Urban'] = company_label.fit_transform(company_data['Urban'])
company_data['US'] = company_label.fit_transform(company_data['US'])
```

```python
company_data
```

```python
company_data.Sales.value_counts()
```

```python
company_data['Sales_Category'] = company_data.Sales.map(lambda x: 1 if x>8 else 0)
```

```python
company_data
```

# 5. Model Building

```python
X = company_data.drop(['Sales','Sales_Category'],axis=1)
y = company_data[['Sales_Category']]
```

```python
X
```

```python
y
```

```python
X.shape,y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=50)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# 6. Model Training

```python
Company_DT = DecisionTreeClassifier(criterion='gini',max_depth=8,min_samples_split=2)
Company_DT.fit(X_train,y_train)
```

# 7. Model Testing

```python
y_pred_train = Company_DT.predict(X_train)
y_pred_test = Company_DT.predict(X_test)
```

```python
y_pred_train
```

```python
y_pred_test
```

# Plotting the Decision Tree

```python
plt.figure(figsize=(16,10))
plot_tree(Company_DT,rounded=True,filled=True)
plt.show()
```
