# 6. Model Building

```python
X = startups_50[['RandD_Spend','Administration','Marketing_Spend','State']]
X
```

```python
y = startups_50[['Profit']]
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

# 7. Model Training 

```python
linear_model = LinearRegression()
linear_model.fit(X_train,y_train) # Model Training
```

```python
print('Slope :',linear_model.coef_)
print('Intercept: ',linear_model.intercept_)
```

# 8. Model Testing

```python
X_test
```

```python
y_test
```

```python
y_pred  = linear_model.predict(X_test)
y_pred
```

# 9. Model Evaluation

```python
print('Mean Squared Error:',mean_squared_error(y_test,y_pred),'\n','R2 Score:',r2_score(y_test,y_pred))
```

# 10. Model Optimization

```python
m1 = sfa.ols('Profit ~ RandD_Spend + Administration + Marketing_Spend + State',data = startups_50).fit()
```

```python
m1.summary()

#Null Hypothesis: HP is not significant feature for predicting MPG
#Alt  Hypothesis: HP is significant.

#Threshold as 0.05. So if p<0.05 --> reject the null hypothesis
```

```python
m2 = sfa.ols('Profit ~ RandD_Spend + Administration + Marketing_Spend',data = startups_50).fit()
m2.summary()
```

### OBSERVATION

1. m1 ->> RandD_Spend + Administration + Marketing_Spend + State --> Adj_R2 = 0.946 AIC = 1061
2. m2 ->> RandD_Spend + Administration + Marketing_Spend     --> Adj_R2 = 0.948, AIC = 1059

adj r2 score and aic of model 1 is better. so we choose model 1
